# Section 2a: Linear Regression with Single Variable

Now it's time to implement and train a model using one attribute as feature - <code>mileage</code>, and one as target - <code>price</code>. <code>mileage</code> is a numerical colum with integer values from 1 to 177644. 

I will split the data into a training set (80%) and a test set (20%).

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = df_ford[['mileage']]
y = df_ford['price']

# split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
len(X_train)
```

```python
len(X_test)
```

Now it is time to use the train data to fit a model.

```python
ford_regression = LinearRegression()
ford_regression_model = ford_regression.fit(X_train, y_train)
ford_regression_model
```

```python
# Use the model to predict prices for the test set
predictions = ford_regression_model.predict(X_test)
predictions[0:30]
# ford_regression_model.predict(np.array([[4000]]))[0]
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from prettytable import PrettyTable

mileage_r2 = r2_score(y_test, predictions)
mileage_mse = mean_squared_error(y_test, predictions)
mileage_coef = ford_regression_model.coef_ 
mileage_intercept = ford_regression_model.intercept_

ford_single_variable_regression = PrettyTable(["Y","X", "Coefficient", "Intercept", "R Square Score", "Mean Squared Error"])

ford_single_variable_regression.add_row(["price", "mileage", mileage_coef, mileage_intercept, mileage_r2, mileage_mse])

print(ford_single_variable_regression)
```

To fix this issue I'd like to try scaling the data as suggested in the Assignment descritpion. I learned it is used to scale numerical inputs (which might have different scales and ranges) to a standard range and that it is a recommended step in data processing for machine learning algorithms, especially those that use a weighted sum of input variables as explained in <a href="https://machinelearningmastery.com/standardscaler-and-minmaxscaler-transforms-in-python/">"How to Use StandardScaler and MinMaxScaler Transforms in Python"</a>  article. Since in my dataset there are big differences in ranges and scales of the numerical values, I hope scaling will improve the accuracy. For that, I will use a model with the highest accuracy and if I'll be able to improve it, I'll apply this technique to other models as well. 

I will use Python's <code>StandardScaler</code>, which according to its documentation standardizes features by removing the mean and scaling to unit variance.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = df_ford[['year']]
y = df_ford['price']

X = scaler.fit_transform(X)

# split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# train a model using the training data
ford_regression = LinearRegression()
ford_regression_model = ford_regression.fit(X_train, y_train)

# Use the model to predict prices for the test set
predictions = ford_regression_model.predict(X_test)

year_coef = ford_regression_model.coef_ 
year_intercept = ford_regression_model.intercept_

year_score = ford_regression_model.score(X_test, y_test)
year_r2 = r2_score(y_test, predictions)
year_mse = mean_squared_error(y_test, predictions)

table = PrettyTable(["Y","X", "Coefficient", "Intercept", "R Square Score", "Mean Squared Error"])

table.add_row(["price", "year", year_coef, year_intercept, r2_score(y_test, predictions), year_mse])

print(table)
```

As we can see above, scaling input variables using <code>StandardScaler()</code> did not increase accuracy, and even decreased it a little (by ~1%). Maybe in this case, the different scales and ranges of numerical features dont't influence, or at least don't reduce, the model's performance. Or maybe the scaling technique would perform better on a different type of algorithm, like multilayer perceptron.

# Section 2b:  Linear Regression with Multiple Variables

To implement a model using linear regression with multiple variables, I'll first use all the attributes I analyzed in the previous section - <code>year</code>, <code>mileage</code>, <code>engineSize</code>, and <code>mpg</code>.

```python
X = df_ford[['year', 'mileage', 'engineSize', 'mpg']]
y = df_ford['price']

# split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# train a model using the training data
ford_regression_mult = LinearRegression()
ford_regression_model_mult = ford_regression_mult.fit(X_train, y_train)

# Use the model to predict prices for the test set
predictions = ford_regression_model_mult.predict(X_test)

mult_score = ford_regression_model_mult.score(X_test, y_test)
mult_r2 = r2_score(y_test, predictions)
mult_mse = mean_squared_error(y_test, predictions)

ford_multiple_variable_regression_table = PrettyTable(["Y","X", "R Square Score", "Mean Squared Error"])

ford_multiple_variable_regression_table.add_row(["price", ["mileage","year","engSize","mpg"], mult_r2, mult_mse ])

print(ford_multiple_variable_regression_table)
```

As we can see, the model using multiple variables reached siginificantly higher accuracy that models using each one of the features separately. It is about 74%.

I now would like to see whether adding more features will increase the model's accuracy even further. To do so, I will need to convert categorical variables into numerical values. I will use <code>model</code>, <code>transmission</code>, and <code>fuelType</code>. 

```python
dummies = df_ford.copy()
dummies = pd.get_dummies(dummies, columns=['model'])
dummies = pd.get_dummies(dummies, columns=['transmission'])
dummies = pd.get_dummies(dummies, columns=['fuelType'])

dummies.head()
```

Categorical columns now have numeric values.

```python
X = dummies.drop('price', axis=1)

y = dummies['price']

# split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# train a model using the training data
ford_regression_mult2 = LinearRegression()
ford_regression_model_mult2 = ford_regression_mult.fit(X_train, y_train)

# Use the model to predict prices for the test set
predictions = ford_regression_model_mult2.predict(X_test)

mult_score = ford_regression_model_mult2.score(X_test, y_test)
mult_r2 = r2_score(y_test, predictions)
mult_mse = mean_squared_error(y_test, predictions)

ford_multiple_variable_regression_table = PrettyTable(["Y","X", "R Square Score", "Mean Squared Error"])

ford_multiple_variable_regression_table.add_row(["price", "All remaining columns", mult_r2, mult_mse])

print(ford_multiple_variable_regression_table)
```

As we can see in the table above, adding more features further increased the accuracy of the model. It is now ~85%. 

I'd like to see if the accuracy will improve if I apply log transformation like I did in the previous section. 

```python
X = dummies.drop('price', axis=1)

y = dummies['price']


y_trans = np.log1p(y)

# split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_trans, test_size=0.2)

# train a model using the training data
ford_regression_mult2 = LinearRegression()
ford_regression_model_mult2 = ford_regression_mult.fit(X_train, y_train)

# Use the model to predict prices for the test set
predictions = ford_regression_model_mult2.predict(X_test)

mult_score = ford_regression_model_mult2.score(X_test, y_test)
mult_r2 = r2_score(y_test, predictions)
mult_mse = mean_squared_error(y_test, predictions)

ford_multiple_variable_regression_table = PrettyTable(["Y","X", "R Square Score", "Mean Squared Error"])

ford_multiple_variable_regression_table.add_row(["price", "All remaining columns", mult_r2, mult_mse])

print(ford_multiple_variable_regression_table)
```

# Section 4: Classification with Logistic Regression

In this section, I'd like to use Logistic Regression algorithm. I want to learn how it is implemented in scikit learn, what parameters it uses, and and how they can influence model's performance.

Because my target variable <code>custcat</code> has 4 possible values, I'll perform multiclass classification.

I'll strat with Logistic Regression seems to be the one of the most popular classifiers for begginers. As mentioned above, I will apply the algorithm to a multiclass problem (predicting one of four customer categories) - a modification of default binary logistic regression. I'll do so by setting one of the model's parameters - <code>multi_class</code>, to <code>multinominal</code>.

```python
X = df_customers.drop('custcat', axis=1)
y = df_customers['custcat']
```

```python
# split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
customer_category_model_lr = LogisticRegression(solver='lbfgs', multi_class='multinomial', max_iter=10000)
```

```python
customer_category_model_lr.fit(X_train, y_train)
```

```python
customer_category_model_lr.score(X_test, y_test)
```

As we can see, the model accuracy is not very high - about 40%. I hope it can be improved.

While reading more about <code>LogisticRegression()</code> parameters, I learned that one of the useful ones for multiclass problems is <code>penalty</code> responsible for regularization. It forces the model to seek smaller weights, and that might lead to better performance on unseen data.

According to documentation, parameter <code>C</code> can be used to set the weight in the penalty and "smaller values specify stronger regularization".

```python
print("Accuracy of the model with no penalty: %f\n"%(customer_category_model_lr.score(X_test, y_test)))

print("Accuracy of models using penalty and different C values: ")
for c in [0.00000001, 0.0000001, 0.000001, 0.00001, 0.0001, 0.001, 0.01, 0.1, 1.0]:
    customer_category_model_lr = LogisticRegression(solver='lbfgs', multi_class='multinomial', max_iter=10000, penalty='l2', C=c)
    customer_category_model_lr.fit(X_train, y_train)
    print("c=%.8f, accuracy=%f"%(c, customer_category_model_lr.score(X_test, y_test)))
```

The highest accuracy (~41%) was reached by the model with the <code>C</code> parameter set to 0.1000000, which is a very small improvement from the model with no penalty whose accuracy is ~40%. I don't think it is significant increase, and with a relatively small dataset I don't think that based on this test I can rely on this method to guarantee an improvement.

The highest <code>C</code> value led to the accuracy equal to that of the model using no penalty. 

Now, to see if I can more improve the accuracy of the model, I'd like to try data scaling like in the regression task.
